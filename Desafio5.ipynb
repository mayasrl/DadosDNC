```python
import pandas as pd
```

```python
file_path = '/content/data.csv'
```

```python
try:
    data = pd.read_csv(file_path)
except UnicodeDecodeError:
    data = pd.read_csv(file_path, encoding='ISO-8859-1')

print("Visualização inicial do dataset:")
print(data.head())
```

```python
print("\nDescrição do dataset:")
print(data.describe(include='all'))

print("\nInformações sobre os tipos de dados:")
print(data.info())
```

```python
print("Verificando valores nulos em cada coluna:")
print(data.isna().sum())
```

```python
data_cleaned = data.dropna()
```

```python
print("\nApós remover valores nulos:")
print(data_cleaned.isna().sum())
print(f"Quantidade de registros restantes: {len(data_cleaned)}")
```

```python
print("Nomes das colunas no dataset:")
print(data_cleaned.columns)
```

```python
print("Verificando valores inválidos na coluna de preços unitários (UnitPrice):")
print(data_cleaned[data_cleaned['UnitPrice'].isna() | (data_cleaned['UnitPrice'] <= 0)])
```

```python
data_cleaned = data_cleaned[data_cleaned['UnitPrice'] > 0]
```

```python
print("\nVerificando valores inválidos na coluna de quantidade (Quantity):")
print(data_cleaned[data_cleaned['Quantity'].isna() | (data_cleaned['Quantity'] <= 0)])
```

```python
data_cleaned = data_cleaned[data_cleaned['Quantity'] > 0]
```

```python
print("\nDataset final após os filtros:")
print(data_cleaned.head())
print(f"Quantidade de registros restantes: {len(data_cleaned)}")
```

```python
print("Verificando linhas duplicadas no dataset:")
duplicated_rows = data_cleaned.duplicated()
print(f"Número de linhas duplicadas: {duplicated_rows.sum()}")
```

```python
data_cleaned = data_cleaned.drop_duplicates()
```

```python
print("\nApós remoção de duplicatas:")
print(f"Número de registros restantes: {len(data_cleaned)}")
print(f"Número de linhas duplicadas restantes: {data_cleaned.duplicated().sum()}")
```

```python
data_cleaned['InvoiceNo'] = data_cleaned['InvoiceNo'].astype(str)
data_cleaned['StockCode'] = data_cleaned['StockCode'].astype(str)
data_cleaned['Description'] = data_cleaned['Description'].astype(str)
data_cleaned['Country'] = data_cleaned['Country'].astype(str)
```

```python
data_cleaned['Quantity'] = data_cleaned['Quantity'].astype(int)
```

```python
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'], errors='coerce')
```

```python
data_cleaned['UnitPrice'] = data_cleaned['UnitPrice'].astype(float)
```

```python
data_cleaned['CustomerID'] = pd.to_numeric(data_cleaned['CustomerID'], errors='coerce').astype('Int64')
```

```python
print("Tipos de dados após a correção:")
print(data_cleaned.dtypes)
```

```python
print("\nExemplo de dados corrigidos:")
print(data_cleaned.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
outliers = data_cleaned[(data_cleaned['Quantity'] > 10000) & (data_cleaned['UnitPrice'] > 5000)]

print(f"\nOutliers encontrados:\n{outliers}")
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(x=data_cleaned['Quantity'])
plt.title('Boxplot - Quantity')
```

```python
plt.subplot(1, 2, 2)
sns.boxplot(x=data_cleaned['UnitPrice'])
plt.title('Boxplot - UnitPrice')

plt.tight_layout()
plt.show()
```

```python
data_cleaned = data_cleaned[~((data_cleaned['Quantity'] > 10000) & (data_cleaned['UnitPrice'] > 5000))]

print(f"\nNúmero de registros após remoção de outliers: {len(data_cleaned)}")
```

```python
data_cleaned['TotalPrice'] = data_cleaned['Quantity'] * data_cleaned['UnitPrice']

print("\nExemplo de dados com a nova coluna 'TotalPrice':")
print(data_cleaned[['Quantity', 'UnitPrice', 'TotalPrice']].head())
```

```python
ultima_data = data_cleaned['InvoiceDate'].max()

print(f"\nData da última compra: {ultima_data}")
```

```python
country_sales = data_cleaned.groupby('Country').agg(TotalSales=('TotalPrice', 'sum')).reset_index()
top_10_countries = country_sales.sort_values(by='TotalSales', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='TotalSales', y='Country', data=top_10_countries, palette='viridis')
plt.title('Top 10 Países com Maior Valor em Vendas')
plt.xlabel('Valor Total de Vendas')
plt.ylabel('País')
plt.show()
```

```python
product_sales = data_cleaned.groupby('Description').agg(TotalQuantity=('Quantity', 'sum')).reset_index()
top_10_products = product_sales.sort_values(by='TotalQuantity', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='TotalQuantity', y='Description', data=top_10_products, palette='Blues')
plt.title('Top 10 Produtos Mais Vendidos')
plt.xlabel('Quantidade Total Vendida')
plt.ylabel('Produto')
plt.show()
```

```python
data_cleaned['Month'] = data_cleaned['InvoiceDate'].dt.to_period('M')
```

```python
monthly_sales = data_cleaned.groupby('Month').agg(TotalSales=('TotalPrice', 'sum')).reset_index()
```

```python
# Verificar valores não numéricos ou NaN
print(monthly_sales['TotalSales'].isna().sum())  # Verificar quantos valores NaN
print(monthly_sales['TotalSales'].dtype)  # Verificar o tipo da coluna
```

```python
# Substituir NaN por 0 ou outro valor de sua escolha
monthly_sales['TotalSales'] = monthly_sales['TotalSales'].fillna(0)

# Garantir que a coluna é numérica
monthly_sales['TotalSales'] = pd.to_numeric(monthly_sales['TotalSales'], errors='coerce')
```

```python
print(monthly_sales['Month'].dtype)  # Verificar o tipo de dados da coluna 'Month'
```

```python
monthly_sales['Month'] = monthly_sales['Month'].astype(str)  # Converte para string, se necessário
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='TotalSales', data=monthly_sales, marker='o', color='green')
plt.title('Valor de Venda Total por Mês')
plt.xlabel('Mês')
plt.ylabel('Valor Total de Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_country_sales = data_cleaned[data_cleaned['Country'].isin(top_10_countries['Country'])]
```

```python
monthly_country_sales = top_10_country_sales.groupby(['Month', 'Country']).agg(TotalSales=('TotalPrice', 'sum')).reset_index()
```

```python
# Verificar valores não numéricos ou NaN em TotalSales
print(monthly_country_sales['TotalSales'].isna().sum())  # Verificar quantos valores NaN
print(monthly_country_sales['TotalSales'].dtype)  # Verificar tipo da coluna
```

```python
# Substituir NaN por 0 ou outro valor de sua escolha
monthly_country_sales['TotalSales'] = monthly_country_sales['TotalSales'].fillna(0)

# Garantir que a coluna é numérica
monthly_country_sales['TotalSales'] = pd.to_numeric(monthly_country_sales['TotalSales'], errors='coerce')
```

```python
print(monthly_country_sales['Month'].dtype)  # Verificar tipo da coluna 'Month'
```

```python
monthly_country_sales['Month'] = monthly_country_sales['Month'].astype(str)  # Se necessário, converte para string
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Month', y='TotalSales', hue='Country', data=monthly_country_sales)
plt.title('Valor de Venda Total por Mês e por País (Top 10 Países)')
plt.xlabel('Mês')
plt.ylabel('Valor Total de Vendas')
plt.xticks(rotation=45)
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'])
last_purchase_date = data_cleaned['InvoiceDate'].max()

recency = data_cleaned.groupby('CustomerID')['InvoiceDate'].max()
recency = (last_purchase_date - recency).dt.days

frequency = data_cleaned.groupby('CustomerID')['InvoiceNo'].nunique()

monetary = data_cleaned.groupby('CustomerID')['TotalPrice'].sum()
monetary = monetary / frequency
rfm = pd.DataFrame({
    'Recency': recency,
    'Frequency': frequency,
    'Monetary': monetary
})

print(rfm.head())
```
