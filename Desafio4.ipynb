```python
sns.pairplot(base_mkt, x_vars=['youtube', 'facebook', 'newspaper'], y_vars=['sales'])
```

```python
import matplotlib.pyplot as plt

plt.subplot(2,2,1)
sns.histplot(base_mkt["youtube"]).set(title="Youtube")
plt.subplot(2,2,2)
sns.histplot(base_mkt["facebook"]).set(title="Facebook")
plt.subplot(2,2,3)
sns.histplot(base_mkt["newspaper"]).set(title="Newspaper")
plt.tight_layout()
```

```python
X = base_mkt[['youtube', 'facebook', 'newspaper']]

Y= base_mkt[['sales']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7,  test_size = 0.3,  random_state =40)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train , Y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r = r2_score(Y_test, y_pred)
```

```python
print("r_quadrado:",r )
```
